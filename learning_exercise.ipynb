## Now we will make a circuit with two qubits and add in a two qubit "controlled" gate discussed in the lecture

```python
def create_quantum_circuit_two_qubit(
    input_state_q1 = [0,1],
    input_state_q2 = [0,1]
    ):
    circuit = QuantumCircuit(2)
    circuit.initialize(input_state_q1, 0)
    circuit.initialize(input_state_q2, 1)
    sim = Aer.get_backend('qasm_simulator')  # Define our simulation method
    ### In this section, add code that creates a:
    ### 1) controlled not (controlled x) or cx
    ###    where the first qubit is the control and the
    ###    second is the target
    circuit.cx(0,1)
    ###
    circuit.save_statevector() # Tell the simulation to save the state vector as we run it
    result = sim.run(circuit, shots = 1000).result()
    out_vector = result.get_statevector()
    counts = result.get_counts(circuit)
    return {'outvector': out_vector,
            'counts': counts}
```

```python
### Example starting states
super_position_state = [1/math.sqrt(2), 1j/math.sqrt(2)] # superposition state
ground_state = [1,0]
excited_state = [0,1]

### Run the following for each different gate in the above block
### Choose from the above states which to start with to see the gate action
out_vector = create_quantum_circuit_two_qubit(
    input_state_q1 = excited_state,  # Change the states
    input_state_q2 = ground_state  # Change the states
    )['outvector']
plot_bloch_multivector(out_vector)
```

```python
counts = create_quantum_circuit_two_qubit(
    input_state_q1 = ground_state,  # Change the states
    input_state_q2 = ground_state  # Change the states
    )['counts']
plot_histogram(counts)
```
