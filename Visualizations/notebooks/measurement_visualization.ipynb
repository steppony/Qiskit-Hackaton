```python
from qiskit import QuantumCircuit,QuantumRegister, ClassicalRegister, Aer, execute
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(2,2)
qc.measure(0,0)
qc.measure(1,1)

result = execute(qc, Aer.get_backend('qasm_simulator')).result()
plot_histogram(result.get_counts(qc))
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.h(1)
qc.measure(0,0)
qc.measure(1,1)

result = execute(qc, Aer.get_backend('qasm_simulator')).result()
plot_histogram(result.get_counts(qc))
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.h(1)
qc.x(0)
qc.measure(0,0)
qc.measure(1,1)

result = execute(qc, Aer.get_backend('qasm_simulator')).result()
plot_histogram(result.get_counts(qc))
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.h(1)
qc.x(0)
qc.cz(1,0)
qc.measure(0,0)
qc.measure(1,1)

result = execute(qc, Aer.get_backend('qasm_simulator')).result()
plot_histogram(result.get_counts(qc))
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.h(1)
qc.x(0)
qc.cz(1,0)
qc.measure(0,0)
qc.measure(1,1)

result = execute(qc, Aer.get_backend('qasm_simulator'), shots=1).result()
plot_histogram(result.get_counts(qc))
```

```python
qr = QuantumRegister(2, name='q')
cr = ClassicalRegister(2, name='output')
qc = QuantumCircuit(qr, cr)
qc.h(0)
qc.h(1)
qc.x(0)
qc.cz(1,0)
qc.measure(0,0)
qc.measure(1,1)
qc.draw('mpl')
```
