```python
from qiskit import QuantumCircuit,QuantumRegister, ClassicalRegister
```

```python
a = QuantumRegister(2, name="a")  
b = QuantumRegister(1, name="b")
cr1 = ClassicalRegister(1, name="outcome1")
cr2 = ClassicalRegister(1, name="outcome2") 
teleportation_circuit = QuantumCircuit(a, b, cr1, cr2)
```

```python
def psi_00(circuit, q1, q2):
    circuit.h(q1)
    circuit.cnot(q1, q2)

def Bell_measurement(circuit, q1, q2, cr1, cr2):
    circuit.cnot(q2, q1)
    circuit.h(q2)
    circuit.measure(q1, cr1)
    circuit.measure(q2, cr2)

def conditional_gates(circuit, q0, q1, q2, cr1, cr2):
    circuit.x(q0).c_if(cr1, 1)
    circuit.z(q0).c_if(cr2, 1)
```

```python
#STEP 1:  Create the Bell state between q0 and q1
psi_00(teleportation_circuit, 0, 1)

#STEP 2: Bell Measurement
teleportation_circuit.barrier()
Bell_measurement(teleportation_circuit, 1, 2, cr1, cr2)

#STEP 3: Conditional Gates
teleportation_circuit.barrier()
conditional_gates(teleportation_circuit, 0, 1, 2, cr1, cr2)

teleportation_circuit.draw('mpl', scale=5)
```
